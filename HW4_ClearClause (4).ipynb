# Part A: Multi-Document Parser (20 points)

## What this part is about

Before the rule engine can evaluate anything, your system needs to understand the structure of what it received. In the real world, a legal AI system does not receive a single clean contract, it receives a package of documents, and those documents interact with each other in ways that matter enormously for risk assessment.

This part builds two deterministic (no LLM) components. The first is a **document classifier** that reads a document and identifies what type it is: an original agreement, an amendment that modifies an original, or a schedule that adds scope and special terms. Classification also extracts which clause types the document covers and whether the document makes any explicit claim about controlling other documents.

The second is a **conflict detector**. When two documents both address the same clause, say both have payment terms, your detector compares the values and determines which document controls. This is a non-trivial problem because the documents in this package make conflicting precedence claims. The Schedule says it controls over the MSA and any Amendment on payment and IP matters. The Amendment says it supersedes the MSA on payment terms. When both documents claim control over the same clause, the right answer is not to pick one arbitrarily, it is to flag the ambiguity with low confidence and explain why.

## The three conflicts you must find

Read the document package carefully and identify these before coding:

1. **Payment terms** appear in all three documents with different values (60, 90, and 120 days), and both the Amendment and the Schedule claim their value controls. This is genuinely ambiguous. Your detector must flag it as low-confidence resolution.

2. **IP ownership** is granted to the Client by the MSA, then partially reversed by the Schedule for ML models trained on Client data. A system that only reads the MSA will miss this entirely.

3. **Termination rights** are symmetric in the MSA (neither party can terminate for convenience), but the Amendment gives the Vendor unilateral 7-day termination. This is a significant power shift that only becomes visible when you compare the two documents.

## Why the homework uses deterministic classification

As noted in the architecture slide, document classification is genuinely a neural task in production. A fine-tuned model handles the linguistic variation that keyword rules cannot. But in this homework you will implement the classifier with keyword heuristics, and the reason is architectural rather than a statement about what is technically possible.

The classifier feeds the conflict detector, which feeds the rule engine, which produces findings that must be fully auditable. A stochastic step anywhere in that chain means two runs on the same document could produce different risk reports, and you cannot explain which one is correct. For this homework, deterministic classification keeps that audit trail clean and lets you focus on the harder design challenges in the layers above it. In a production system, the equivalent guarantee comes from confidence thresholds and human review for low-confidence predictions, not from removing the model entirely.

**Grading rubric:** `DocumentClassifier` correctly identifies all three types (5 pts). `ConflictDetector` finds all three conflicts (8 pts). Precedence logic is internally consistent and handles the ambiguous three-way payment conflict with `LOW` confidence (7 pts).

````python
# Part A: Step 1: Document Classifier
#
# The classifier uses keyword and structural heuristics to determine what kind
# of legal document it is reading. Think about what distinguishes each type:
#
#   MSA/Agreement: typically has numbered sections covering many topics,
#                  defines both parties, uses phrases like 'agrees to provide'
#   Amendment:     explicitly references the original agreement by name and date,
#                  uses phrases like 'supersedes', 'replaces', 'deleted in its entirety'
#   Schedule:      titled as 'Schedule', 'Exhibit', or 'Addendum',
#                  may contain 'notwithstanding' clauses that override the main agreement
#
# class DocumentMeta:
#     doc_id: str
#     doc_type: str                  # 'MSA', 'Amendment', 'Schedule', 'Unknown'
#     effective_date: str | None     # date string as it appears in the document
#     parties: list[str]             # extracted party names
#     precedence_claims: list[str]   # full sentences claiming this doc controls others
#                                    # look for: 'shall control', 'supersedes', 'notwithstanding'
#     clauses_covered: list[str]     # which clause types this document addresses
#                                    # use this fixed vocabulary:
#                                    # 'payment_terms', 'ip_ownership', 'governing_law',
#                                    # 'confidentiality', 'termination', 'indemnification',
#                                    # 'liability_cap', 'dispute_resolution',
#                                    # 'non_compete', 'subcontractor'
#
# Implementation note: clauses_covered should be populated by checking for
# the presence of keywords associated with each clause type in the document text.
# Define your keyword map explicitly. This is one of the design decisions
# that is your work, not something an AI tool should generate for you.

from dataclasses import dataclass, field
from typing import Optional
import re

# TODO: Define DocumentMeta dataclass

@dataclass
class DocumentMeta:
    doc_id: str
    doc_type: str
    effective_date: Optional[str]
    parties: list[str]
    precedence_claims: list[str]
    clauses_covered: list[str]

# TODO: Implement DocumentClassifier

CLAUSE_KEYWORD_MAP: dict[str, list[str]] = {
    "payment_terms": [
        "payment", "invoice", "invoices", "net 30", "net-30",
        "days of receipt", "days of milestone", "late payment",
        "suspension grace period", "non-payment",
    ],
    "ip_ownership": [
        "intellectual property", "work product", "deliverables",
        "exclusive property", "ownership", "ml models", "trained on client data",
        "commercial product offerings", "pre-existing tools",
    ],
    "governing_law": [
        "governing law", "laws of the state", "construed in accordance",
        "conflict of law",
    ],
    "confidentiality": [
        "confidential information", "hold in confidence", "confidentiality",
        "nondisclosure", "non-disclosure",
    ],
    "termination": [
        "terminate", "termination", "non-renewal", "auto-renew",
        "written notice", "terminate for cause", "terminate for convenience",
    ],
    "indemnification": [
        "indemnify", "indemnification", "hold harmless",
        "third-party claims", "negligence", "willful misconduct",
    ],
    "liability_cap": [
        "limitation of liability", "total liability", "shall not exceed",
        "aggregate liability", "indirect", "consequential damages", "incidental",
    ],
    "dispute_resolution": [
        "arbitration", "jams", "mediation", "dispute",
        "comprehensive arbitration rules", "binding arbitration",
    ],
    "non_compete": [
        "non-compete", "noncompete", "solicit", "non-solicit",
        "competing", "restraint of trade",
    ],
    "subcontractor": [
        "subcontractor", "subcontractors", "subcontract",
        "third-party vendor", "engage subcontractors",
    ],
}


class DocumentClassifier:

    PRECEDENCE_KEYWORDS = [
        "shall control", "supersedes", "notwithstanding",
        "deleted in its entirety", "replaces",
    ]

    DATE_PATTERNS = [
        r'\b(?:January|February|March|April|May|June|July|August|September|'
        r'October|November|December)\s+\d{1,2},\s+\d{4}\b',
        r'\b\d{1,2}(?:st|nd|rd|th)\s+day\s+of\s+'
        r'(?:January|February|March|April|May|June|July|August|September|'
        r'October|November|December),?\s+\d{4}\b',
        r'\b\d{4}-\d{2}-\d{2}\b',
    ]

    PARTY_PATTERN = re.compile(
        r'(?:between|by and between|entered into by)\s+'
        r'(.{5,60}?)'
        r'\s+and\s+'
        r'(.{5,60}?)'
        r'(?=\s*[\(\,\.])',
        re.IGNORECASE
    )

    def classify(self, doc_id: str, text: str) -> DocumentMeta:
        return DocumentMeta(
            doc_id=doc_id,
            doc_type=self._detect_type(text),
            effective_date=self._extract_date(text),
            parties=self._extract_parties(text),
            precedence_claims=self._extract_precedence_claims(text),
            clauses_covered=self._detect_clauses(text),
        )

    def _detect_type(self, text: str) -> str:
        t = text.lower()

        # Check Schedule FIRST — before Amendment, because Schedule text
        # often references "amendment" in its body
        schedule_signals = ["schedule", "exhibit", "addendum"]
        header = t[:300]
        if any(sig in header for sig in schedule_signals):
            return "Schedule"

        amendment_signals = [
            "amendment", "amended and restated", "as amended",
            "hereby amends", "deleted in its entirety", "replaces section",
        ]
        if any(sig in t for sig in amendment_signals):
            return "Amendment"

        msa_signals = [
            "agrees to provide", "master service", "master agreement",
            "services agreement", "statement of work",
        ]
        has_numbered_sections = bool(re.search(r'\n\s*\d+[\.\)]\s+\w', text))
        if any(sig in t for sig in msa_signals) or has_numbered_sections:
            return "MSA"

        return "Unknown"

    def _extract_date(self, text: str) -> Optional[str]:
        effective_pattern = re.compile(
            r'effective(?:\s+date)?(?:\s*:\s*|\s+as\s+of\s+)'
            r'(' + '|'.join(self.DATE_PATTERNS) + r')',
            re.IGNORECASE
        )
        m = effective_pattern.search(text)
        if m:
            return m.group(1).strip()
        for pat in self.DATE_PATTERNS:
            m = re.search(pat, text, re.IGNORECASE)
            if m:
                return m.group(0).strip()
        return None

    def _extract_parties(self, text: str) -> list[str]:
        m = self.PARTY_PATTERN.search(text)
        if not m:
            return []
        parties = []
        for raw in [m.group(1), m.group(2)]:
            cleaned = re.sub(
                r',?\s*a\s+\w[\w\s]+(?:corporation|company|llc|inc)\.?$',
                '', raw, flags=re.IGNORECASE
            )
            cleaned = cleaned.strip(' ,.()')
            if cleaned:
                parties.append(cleaned)
        return parties

    def _extract_precedence_claims(self, text: str) -> list[str]:
        sentences = re.split(r'(?<=[.!?])\s+', text)
        return [
            s.strip() for s in sentences
            if any(kw in s.lower() for kw in self.PRECEDENCE_KEYWORDS)
        ]

    def _detect_clauses(self, text: str) -> list[str]:
        t = text.lower()
        return [
            clause for clause, keywords in CLAUSE_KEYWORD_MAP.items()
            if any(kw.lower() in t for kw in keywords)
        ]
````

````python
# Part A: Step 2: Conflict Detector
#
# A conflict exists when two documents both have a non-None value for the same
# ContractFields field and those values differ. Once a conflict is found,
# the detector must decide which document's value controls, and must be
# honest about how confident that determination is.
#
# Precedence logic you must implement (apply in this order):
#   1. A document with an explicit 'shall control' or 'notwithstanding' claim
#      over a specific clause wins over one that does not claim control,
#      but only for the specific clauses claimed, not for all clauses.
#   2. An Amendment supersedes the MSA for clauses it explicitly replaces.
#   3. When two documents both claim control over the same clause
#      (the payment terms three-way conflict is this case),
#      resolution_confidence must be 'LOW'.
#      The rationale must explain the ambiguity rather than hiding it.
#   4. risk_amplified must be True when the controlling value produces
#      a worse outcome for the Client than the non-controlling value.
#      For example: if the controlling payment term is 120 days and the
#      non-controlling value was 60 days, the conflict made risk worse.
#
# class ConflictReport:
#     conflicts: list[dict]   # each dict:
#                             #   'clause': str
#                             #   'doc_a_value': any
#                             #   'doc_b_value': any
#                             #   'controlling_doc': str
#                             #   'controlling_value': any
#                             #   'rationale': str
#                             #   'risk_amplified': bool
#     has_conflicts: bool
#     resolution_confidence: str   # 'HIGH', 'MEDIUM', or 'LOW'

# TODO: Define ConflictReport dataclass
@dataclass
class ConflictReport:
    conflicts: list[dict]
    has_conflicts: bool
    resolution_confidence: str

# TODO: Implement ConflictDetector
class ConflictDetector:

    def detect(
        self,
        doc_a_meta: DocumentMeta,
        doc_a_fields: dict,
        doc_b_meta: DocumentMeta,
        doc_b_fields: dict,
    ) -> ConflictReport:
        conflicts = []
        overall_confidence = "HIGH"

        shared_clauses = set(doc_a_fields.keys()) & set(doc_b_fields.keys())

        for clause in shared_clauses:
            val_a = doc_a_fields[clause]
            val_b = doc_b_fields[clause]

            if val_a is None or val_b is None:
                continue
            if val_a == val_b:
                continue

            controlling_doc, controlling_val, rationale, confidence, risk_amp = \
                self._resolve(clause, doc_a_meta, val_a, doc_b_meta, val_b)

            conflicts.append({
                "clause":            clause,
                "doc_a_id":          doc_a_meta.doc_id,
                "doc_a_value":       val_a,
                "doc_b_id":          doc_b_meta.doc_id,
                "doc_b_value":       val_b,
                "controlling_doc":   controlling_doc,
                "controlling_value": controlling_val,
                "rationale":         rationale,
                "risk_amplified":    risk_amp,
                "confidence":        confidence,
            })

            if confidence == "LOW":
                overall_confidence = "LOW"
            elif confidence == "MEDIUM" and overall_confidence == "HIGH":
                overall_confidence = "MEDIUM"

        return ConflictReport(
            conflicts=conflicts,
            has_conflicts=len(conflicts) > 0,
            resolution_confidence=overall_confidence,
        )

    def _resolve(self, clause, meta_a, val_a, meta_b, val_b):
        a_claims = self._claims_control(meta_a, clause)
        b_claims = self._claims_control(meta_b, clause)

        if a_claims and b_claims:
            rationale = (
                f"Both {meta_a.doc_id} ({meta_a.doc_type}) and "
                f"{meta_b.doc_id} ({meta_b.doc_type}) claim to control "
                f"'{clause}'. Precedence is legally ambiguous. "
                f"Legal review required before relying on either value."
            )
            risk_amp = self._is_riskier_for_client(clause, val_a, val_b)
            return meta_b.doc_id, val_b, rationale, "LOW", risk_amp

        if a_claims and not b_claims:
            rationale = (
                f"{meta_a.doc_id} ({meta_a.doc_type}) explicitly claims control "
                f"over '{clause}' via a 'shall control' or 'notwithstanding' clause."
            )
            risk_amp = self._is_riskier_for_client(clause, val_b, val_a)
            return meta_a.doc_id, val_a, rationale, "HIGH", risk_amp

        if b_claims and not a_claims:
            rationale = (
                f"{meta_b.doc_id} ({meta_b.doc_type}) explicitly claims control "
                f"over '{clause}' via a 'shall control' or 'notwithstanding' clause."
            )
            risk_amp = self._is_riskier_for_client(clause, val_a, val_b)
            return meta_b.doc_id, val_b, rationale, "HIGH", risk_amp

        if meta_a.doc_type == "Amendment" and meta_b.doc_type == "MSA":
            rationale = (
                f"{meta_a.doc_id} is an Amendment and supersedes the MSA "
                f"({meta_b.doc_id}) for '{clause}'."
            )
            risk_amp = self._is_riskier_for_client(clause, val_b, val_a)
            return meta_a.doc_id, val_a, rationale, "MEDIUM", risk_amp

        if meta_b.doc_type == "Amendment" and meta_a.doc_type == "MSA":
            rationale = (
                f"{meta_b.doc_id} is an Amendment and supersedes the MSA "
                f"({meta_a.doc_id}) for '{clause}'."
            )
            risk_amp = self._is_riskier_for_client(clause, val_a, val_b)
            return meta_b.doc_id, val_b, rationale, "MEDIUM", risk_amp

        if meta_a.doc_id <= meta_b.doc_id:
            rationale = (
                f"No explicit precedence claim. {meta_b.doc_id} assumed to "
                f"supersede {meta_a.doc_id} as the later document."
            )
            risk_amp = self._is_riskier_for_client(clause, val_a, val_b)
            return meta_b.doc_id, val_b, rationale, "MEDIUM", risk_amp
        else:
            rationale = (
                f"No explicit precedence claim. {meta_a.doc_id} assumed to "
                f"supersede {meta_b.doc_id} as the later document."
            )
            risk_amp = self._is_riskier_for_client(clause, val_b, val_a)
            return meta_a.doc_id, val_a, rationale, "MEDIUM", risk_amp

    def _claims_control(self, meta: DocumentMeta, clause: str) -> bool:
        clause_keywords = {
            "payment_terms": ["payment", "invoice"],
            "ip_ownership":  ["ip", "intellectual property", "ml model", "deliverable"],
            "termination":   ["terminat", "term"],
            "governing_law": ["governing", "law"],
        }
        relevant_kws = clause_keywords.get(clause, [clause.replace("_", " ")])
        for claim in meta.precedence_claims:
            if any(kw in claim.lower() for kw in relevant_kws):
                return True
        return False

    def _is_riskier_for_client(self, clause: str, old_val, new_val) -> bool:
        if clause == "payment_terms":
            try:
                return int(new_val) > int(old_val)
            except (TypeError, ValueError):
                return False
        if clause == "ip_ownership":
            risk_order = {"client": 0, "shared": 1, "vendor": 2}
            return risk_order.get(str(new_val), 1) > risk_order.get(str(old_val), 1)
        if clause == "termination":
            return "vendor" in str(new_val).lower() and "vendor" not in str(old_val).lower()
        return old_val != new_val
````

# Part B: Jurisdiction-Aware Symbolic Rule Engine (25 points)

## What this part is about

The rule engine is the core of the symbolic component in your neuro-symbolic system. It takes a structured record of contract field values and deterministically computes a risk assessment. The same input always produces the same output. There is no LLM involved. A lawyer who reads your rule code should be able to verify each decision without any knowledge of machine learning.

This part introduces two ideas that go beyond a simple flat rule list.

The first is **jurisdiction-aware thresholds**. Contract risk does not exist in a vacuum; it exists relative to the law that governs the contract. A non-compete clause in Texas may be enforceable with reasonable scope. The same clause in California is void by statute (California Business and Professions Code Section 16600 makes most non-competes unenforceable). A rule engine that does not account for jurisdiction will either over-flag risks in Texas or under-flag them in California. Your engine must load the correct threshold set once governing law is known.

The second is **compositional (Tier 2) rules**. Some risks are only visible when multiple clause-level findings are combined. Imagine a contract where the Vendor can terminate with 7 days notice, the indemnification runs only one way against the Client, and the ML models trained on Client data belong to the Vendor. Any one of these clauses, reviewed in isolation, produces a HIGH finding. But the combination is qualitatively worse, it means the Vendor can build commercial products from the Client's data, disclaim liability, and walk away in a week. That is a pattern, not just a collection of individual flags. Tier 2 rules detect these patterns.

## Why the rule engine must be stateful across documents

In the previous version of this homework, the rule engine was a stateless function: give it a `ContractFields` object, get a report back. That works for a single document. Here, the Tier 2 compositional rules need to see findings from the merged, conflict-resolved field set, the combined picture of all three documents. The engine accumulates findings before it can fire pattern rules.

## Jurisdiction background you need to know

You do not need to be a lawyer to implement this, but you do need to understand the legal basis for the threshold differences:

**California payment terms:** The California Prompt Payment Act establishes 30-day payment defaults for many service contracts. A 60-day term is already above the statutory default, making it MEDIUM risk. 75+ days is HIGH.

**California non-competes:** Section 16600 of the California Business and Professions Code renders most employee and contractor non-compete clauses void and unenforceable. Any non-compete in a California-governed contract is HIGH risk regardless of its duration or geographic scope, not because it is enforceable, but because including it may expose the Vendor to legal liability for attempting to enforce an illegal clause.

**California ML model IP:** California's consumer privacy protections (CCPA) create additional obligations when a Vendor uses Client data to train commercial models. Even if the IP clause is contractually valid, the data use may trigger regulatory requirements that the contract does not address.

````python
# Part B: Step 2: JurisdictionRuleSet and Factory
#
# JurisdictionRuleSet is a plain dataclass that holds the numeric thresholds
# and default risk levels that vary by state. The rule engine reads from
# this object rather than having thresholds hardcoded in each rule method.
#
# This separation matters for maintainability: adding a new jurisdiction
# means adding one new factory case, not editing ten rule methods.
#
# Required fields in JurisdictionRuleSet:
#   jurisdiction: str
#   payment_medium_threshold_days: int   # above this = MEDIUM
#   payment_high_threshold_days: int     # above this = HIGH
#   non_compete_base_level: str          # 'LOW', 'MEDIUM', or 'HIGH'
#   trained_model_ip_extra_flag: str | None
#       None for most jurisdictions.
#       For California: a string describing the CCPA-related regulatory risk
#       that should be appended to the R04 explanation.
#
# Threshold values to implement:
#
#   California:
#     payment_medium_threshold_days = 45
#     payment_high_threshold_days   = 75
#     non_compete_base_level        = 'HIGH'   (Cal. B&P Code §16600)
#     trained_model_ip_extra_flag   = 'CCPA risk: Vendor use of Client data
#                                      in commercial models may require Client
#                                      consent under Cal. Civ. Code §1798.100'
#
#   Texas:
#     payment_medium_threshold_days = 60
#     payment_high_threshold_days   = 90
#     non_compete_base_level        = 'MEDIUM'  (enforceable if reasonable)
#     trained_model_ip_extra_flag   = None
#
#   Default (any other jurisdiction or None):
#     payment_medium_threshold_days = 60
#     payment_high_threshold_days   = 90
#     non_compete_base_level        = 'MEDIUM'
#     trained_model_ip_extra_flag   = None

# TODO: Define JurisdictionRuleSet dataclass
@dataclass
class JurisdictionRuleSet:
    jurisdiction: str
    payment_medium_threshold_days: int
    payment_high_threshold_days: int
    non_compete_base_level: str
    trained_model_ip_extra_flag: Optional[str]

# TODO: Implement JurisdictionRuleSetFactory with get(governing_law) classmethod
class JurisdictionRuleSetFactory:

    @classmethod
    def get(cls, governing_law: Optional[str]) -> JurisdictionRuleSet:
        if governing_law and "california" in governing_law.lower():
            return JurisdictionRuleSet(
                jurisdiction="California",
                payment_medium_threshold_days=45,
                payment_high_threshold_days=75,
                non_compete_base_level="HIGH",
                trained_model_ip_extra_flag=(
                    "CCPA risk: Vendor use of Client data in commercial models "
                    "may require Client consent under Cal. Civ. Code §1798.100"
                ),
            )
        if governing_law and "texas" in governing_law.lower():
            return JurisdictionRuleSet(
                jurisdiction="Texas",
                payment_medium_threshold_days=60,
                payment_high_threshold_days=90,
                non_compete_base_level="MEDIUM",
                trained_model_ip_extra_flag=None,
            )
        return JurisdictionRuleSet(
            jurisdiction=governing_law or "Default",
            payment_medium_threshold_days=60,
            payment_high_threshold_days=90,
            non_compete_base_level="MEDIUM",
            trained_model_ip_extra_flag=None,
        )
````

````python
# Part B: Step 3: Two-Tier Rule Engine
#
# TIER 1 RULES
# Each rule is a method that checks one field (or a small group of related fields)
# and returns a finding dict:
#   {'field': 'R01', 'level': 'HIGH', 'explanation': '...'}
# The explanation must be a complete sentence that a non-technical reader can
# understand. It should state what was found and why it is a risk.
# Example: "Payment terms of 120 days exceed the California threshold of 75 days,
#           creating cash flow risk and potential rights of suspension for the Vendor."
#
#   R01: Payment terms (jurisdiction-aware threshold)
#   R02: Liability cap absent or basis is vague
#   R03: IP on custom deliverables owned by Vendor rather than Client
#   R04: IP clause covers ML models trained on Client data (with CCPA flag for CA)
#   R05: Auto-renewal with notice period under 14 days
#   R06: Termination for convenience is Vendor-only
#   R07: Termination notice asymmetry: Vendor notice < 14 days AND Client notice > 30 days
#          This rule detects a specific pattern in the Amendment.
#          It should not fire on the MSA alone.
#   R08: Indemnification runs only against Client
#   R09: Non-compete clause (jurisdiction-aware: always HIGH in California)
#   R10: Subcontractor engagement without Client approval
#   R11: Governing law absent from all documents
#   R12: Unresolved cross-document conflict present
#          When has_unresolved_conflict is True, this rule fires at MEDIUM level.
#          The explanation should note that ambiguous precedence creates legal uncertainty.
#
# TIER 2 COMPOSITIONAL RULES
# These run after all Tier 1 rules. They check which Tier 1 rules fired
# and whether the combination matches a known risk pattern.
# They introduce a new risk level: 'CRITICAL' (above HIGH).
#
#   C01: STRUCTURAL IMBALANCE
#       Fires when: R06 (vendor termination) AND R08 (one-sided indemnity)
#                   AND (R03 OR R04) (IP to vendor) all fire
#       Why this pattern matters: individually, each of these is a HIGH risk.
#       Together, they describe a contract where the Vendor controls termination,
#       bears no indemnification obligations, and owns the work product.
#       The Client has all the obligations and none of the protections.
#       That structural imbalance is qualitatively worse than any single clause.
#
#   C02: DATA SOVEREIGNTY RISK
#       Fires when: R04 (ML model IP to vendor) AND R10 (subcontractor without approval)
#       Why this pattern matters: the Vendor can use Client data to train commercial
#       models AND can give those models (and the underlying data pipeline) to
#       subcontractors the Client has never approved. The Client has no visibility
#       into where its data ends up.
#
#   C03: PAYMENT AND TERMINATION TRAP
#       Fires when: R01 fires HIGH AND R06 fires (vendor short-notice termination)
#       Why this pattern matters: the Client owes extended payment after milestone
#       completion, but the Vendor can terminate in 7 days at any point during
#       that payment window. The Client could be in default on payment for work
#       that was terminated before it could be completed.
#
# generate_report() must return:
#   {'overall_risk': str,          # 'CRITICAL' > 'HIGH' > 'MEDIUM' > 'LOW'
#    'tier1_findings': list[dict],
#    'tier2_findings': list[dict],  # empty list if no patterns fire
#    'jurisdiction': str,
#    'critical_count': int,
#    'high_count': int,
#    'medium_count': int,
#    'rule_version': str}

# TODO: Implement ContractRuleEngine

class ContractRuleEngine:

    RULE_VERSION = "1.0.0"

    def __init__(self, ruleset: JurisdictionRuleSet):
        self.rs = ruleset

    def generate_report(self, fields: ContractFields) -> dict:
        tier1 = self._run_tier1(fields)
        tier2 = self._run_tier2(fields, tier1)

        fired_levels = [f["level"] for f in tier1 + tier2]
        if "CRITICAL" in fired_levels:
            overall = "CRITICAL"
        elif "HIGH" in fired_levels:
            overall = "HIGH"
        elif "MEDIUM" in fired_levels:
            overall = "MEDIUM"
        else:
            overall = "LOW"

        return {
            "overall_risk":    overall,
            "tier1_findings":  tier1,
            "tier2_findings":  tier2,
            "jurisdiction":    self.rs.jurisdiction,
            "critical_count":  fired_levels.count("CRITICAL"),
            "high_count":      fired_levels.count("HIGH"),
            "medium_count":    fired_levels.count("MEDIUM"),
            "rule_version":    self.RULE_VERSION,
        }

    # Tier 1 Rules

    def _run_tier1(self, f: ContractFields) -> list[dict]:
        findings = []
        for rule in [
            self._r01, self._r02, self._r03, self._r04,
            self._r05, self._r06, self._r07, self._r08,
            self._r09, self._r10, self._r11, self._r12,
        ]:
            result = rule(f)
            if result:
                findings.append(result)
        return findings

    def _finding(self, rule_id: str, level: str, explanation: str) -> dict:
        return {"field": rule_id, "level": level, "explanation": explanation}

    def _r01(self, f: ContractFields) -> Optional[dict]:
        days = f.controlling_payment_terms_days or f.payment_terms_days
        if days is None:
            return None
        if days > self.rs.payment_high_threshold_days:
            return self._finding("R01", "HIGH",
                f"Payment terms of {days} days exceed the {self.rs.jurisdiction} "
                f"high-risk threshold of {self.rs.payment_high_threshold_days} days, "
                f"creating cash flow risk and potential suspension rights for the Vendor.")
        if days > self.rs.payment_medium_threshold_days:
            return self._finding("R01", "MEDIUM",
                f"Payment terms of {days} days exceed the {self.rs.jurisdiction} "
                f"medium-risk threshold of {self.rs.payment_medium_threshold_days} days.")
        return None

    def _r02(self, f: ContractFields) -> Optional[dict]:
        if f.liability_cap_present is False:
            return self._finding("R02", "HIGH",
                "No liability cap is present in this agreement. "
                "Without a cap, either party's exposure to damages is unlimited.")
        if f.liability_cap_present and f.liability_cap_basis == "vague":
            return self._finding("R02", "MEDIUM",
                "A liability cap is present but its basis is vague, "
                "making it difficult to quantify the Client's actual exposure.")
        return None

    def _r03(self, f: ContractFields) -> Optional[dict]:
        if f.ip_ownership == "vendor":
            return self._finding("R03", "HIGH",
                "Custom deliverables and work product are owned by the Vendor, "
                "not the Client. The Client may not retain rights to work it funded.")
        return None

    def _r04(self, f: ContractFields) -> Optional[dict]:
        if not f.ip_covers_trained_models:
            return None
        explanation = (
            "The IP clause assigns ML models trained on Client data to the Vendor. "
            "The Vendor may incorporate these models into commercial products."
        )
        if self.rs.trained_model_ip_extra_flag:
            explanation += f" {self.rs.trained_model_ip_extra_flag}"
        return self._finding("R04", "HIGH", explanation)

    def _r05(self, f: ContractFields) -> Optional[dict]:
        if f.auto_renewal and f.auto_renewal_notice_days is not None:
            if f.auto_renewal_notice_days < 14:
                return self._finding("R05", "MEDIUM",
                    f"Auto-renewal is active with only {f.auto_renewal_notice_days} days "
                    f"notice required to cancel, giving the Client very little time to opt out.")
        return None

    def _r06(self, f: ContractFields) -> Optional[dict]:
        if f.termination_for_convenience == "vendor_only":
            return self._finding("R06", "HIGH",
                "Only the Vendor may terminate this agreement for convenience. "
                "The Client has no equivalent right and is locked in unless the Vendor commits a material breach.")
        return None

    def _r07(self, f: ContractFields) -> Optional[dict]:
        v = f.termination_notice_vendor_days
        c = f.termination_notice_client_days
        if v is not None and c is not None:
            if v < 14 and c > 30:
                return self._finding("R07", "HIGH",
                    f"Termination notice is asymmetric: the Vendor may terminate in {v} days "
                    f"while the Client requires {c} days notice. "
                    f"This gives the Vendor significant leverage over the relationship.")
        return None

    def _r08(self, f: ContractFields) -> Optional[dict]:
        if f.indemnification_direction == "client_only":
            return self._finding("R08", "HIGH",
                "Indemnification obligations run only against the Client. "
                "The Vendor bears no indemnification obligations, creating a one-sided risk allocation.")
        return None

    def _r09(self, f: ContractFields) -> Optional[dict]:
        if f.non_compete_years is None:
            return None
        level = self.rs.non_compete_base_level
        if level == "HIGH":
            explanation = (
                f"A non-compete clause of {f.non_compete_years} year(s) is present. "
                f"Under {self.rs.jurisdiction} law (Cal. B&P Code §16600), "
                f"non-compete clauses are void and unenforceable. "
                f"Including this clause may expose the Vendor to legal liability."
            )
        else:
            explanation = (
                f"A non-compete clause of {f.non_compete_years} year(s) with "
                f"{f.non_compete_scope} scope is present. "
                f"Under {self.rs.jurisdiction} law this may be enforceable if reasonable, "
                f"but should be reviewed."
            )
        return self._finding("R09", level, explanation)

    def _r10(self, f: ContractFields) -> Optional[dict]:
        if f.subcontractor_approval_required is False:
            return self._finding("R10", "MEDIUM",
                "The Vendor may engage subcontractors without prior Client approval. "
                "The Client has no visibility into who handles its data or deliverables.")
        return None

    def _r11(self, f: ContractFields) -> Optional[dict]:
        if f.governing_law is None:
            return self._finding("R11", "MEDIUM",
                "No governing law clause was identified across any of the documents. "
                "Without a jurisdiction, the applicable legal thresholds cannot be determined.")
        return None

    def _r12(self, f: ContractFields) -> Optional[dict]:
        if f.has_unresolved_conflict:
            return self._finding("R12", "MEDIUM",
                "At least one cross-document conflict could not be resolved with high confidence. "
                "Legally ambiguous precedence creates uncertainty about which terms actually govern.")
        return None

    # Tier 2 Compositional Rules

    def _run_tier2(self, f: ContractFields, tier1: list[dict]) -> list[dict]:
        fired_ids = {x["field"] for x in tier1}
        findings = []
        for rule in [self._c01, self._c02, self._c03]:
            result = rule(fired_ids, f)
            if result:
                findings.append(result)
        return findings

    def _c01(self, fired: set, f: ContractFields) -> Optional[dict]:
        # STRUCTURAL IMBALANCE: R06 + R08 + (R03 or R04)
        if "R06" in fired and "R08" in fired and ("R03" in fired or "R04" in fired):
            return self._finding("C01", "CRITICAL",
                "STRUCTURAL IMBALANCE: The Vendor controls termination (R06), "
                "bears no indemnification obligations (R08), and owns the work product or "
                "ML models (R03/R04). The Client carries all obligations with none of the "
                "protections. This combination is qualitatively worse than any single clause.")
        return None

    def _c02(self, fired: set, f: ContractFields) -> Optional[dict]:
        # DATA SOVEREIGNTY RISK: R04 + R10
        if "R04" in fired and "R10" in fired:
            return self._finding("C02", "CRITICAL",
                "DATA SOVEREIGNTY RISK: The Vendor owns ML models trained on Client data (R04) "
                "and may engage unapproved subcontractors (R10). "
                "The Client has no visibility into where its data ends up or who can access it.")
        return None

    def _c03(self, fired: set, f: ContractFields) -> Optional[dict]:
      days = f.controlling_payment_terms_days or f.payment_terms_days
      r01_high = (
          days is not None
          and days > self.rs.payment_high_threshold_days
          and "R01" in fired
          )
      if r01_high and "R06" in fired:
        return self._finding(
            "C03", "CRITICAL",
            "PAYMENT AND TERMINATION TRAP: The Client owes extended payment after "
            "milestone completion, but the Vendor may terminate with very short notice "
            "at any point during that payment window. The Client could be in default "
            "on payment for work that was terminated before it could be completed."
        )
        return None
````

# Part C: Neural Extraction with Grounding Verification (15 points)

## What this part is about

The neural extractor is where the LLM does its legitimate work: reading raw contract text and populating structured fields that the rule engine can evaluate. This is the only component in your system that processes unstructured natural language. Everything downstream of it, conflict detection, rule evaluation, agent reasoning, works with structured data.

The central challenge in Part C is **grounding verification**. An LLM asked to extract a payment term from a minimal one-paragraph contract may return 30 as a default because that is common in training data, even if the document says nothing about payment terms at all. This is not a bug in the model, it is doing exactly what it was trained to do: produce a fluent, plausible response. But in a legal system, a hallucinated field value is a serious failure. It can cause the rule engine to miss a risk (if the LLM fills in a safe-looking value for a missing field) or to flag a false risk (if the LLM invents a risky value).

Grounding verification is your defense against this. After the LLM extracts a value for a field, you check whether that value can be traced back to actual text in the document. If it cannot, the field is marked as unverified and treated as absent by the rule engine. The check must be deterministic, no second LLM call, no fuzzy matching library, just string operations on the original document text.

## Document-type-aware extraction

The extractor must handle documents differently based on their type. An MSA is a complete contract: every field either exists or is absent. An Amendment is a delta; it only changes specific fields, and all other fields remain as they were in the MSA. If you ask an LLM to extract all fields from an Amendment and it returns a value for `liability_cap_present`, that value is almost certainly hallucinated, the Amendment says nothing about liability.

Your prompt for Amendment extraction must explicitly instruct the LLM to return null for any field not directly addressed. The `merge_fields` function then applies Amendment values on top of MSA values, only overwriting fields that the Amendment actually changed.

## What grounding means operationally

For a numeric field like `payment_terms_days`, grounding means the exact number appears as a digit string in the document. If the LLM extracted 90 and the document contains the string "90" or "ninety (90)", the field is grounded. If the document does not contain either, it is not grounded.

For an enum field like `ip_ownership`, grounding means at least one keyword from a predefined keyword map appears near a relevant context phrase. You define the keyword map. For `ip_ownership = 'vendor'`, the keywords might be `['shall remain the property of vendor', 'exclusive property of vendor', 'vendor retains']`. The document must contain at least one of these to ground the extracted value.

For a boolean field like `subcontractor_approval_required`, grounding means the document contains a relevant phrase about subcontractors. You define what counts as relevant.

**Grading rubric:** Document-type-aware extraction handles both MSA and Amendment correctly (6 pts). Grounding verification catches at least one hallucination in the adversarial test (5 pts). `merge_fields` applies Amendment deltas correctly and logs all changed fields (4 pts).

````python
# Part C: Step 1: LLM Setup
#
# Copy the GroqLLM wrapper class from Lab 8a. No changes are needed.
# If you use a different model than llama-3.3-70b-versatile, note it
# in your AI usage disclosure and explain why.

import os
from groq import Groq
from langchain_core.language_models.llms import LLM
from langchain_core.callbacks.manager import CallbackManagerForLLMRun
from typing import Any, List, Optional

GROQ_API_KEY = 'urs'

# TODO: Paste GroqLLM class from Lab 8a

# TODO: agent_llm = GroqLLM(model='llama-3.3-70b-versatile')

class GroqLLM(LLM):
    model: str = 'llama-3.3-70b-versatile'
    api_key: str = GROQ_API_KEY

    @property
    def _llm_type(self) -> str:
        return "groq"

    def _call(
        self,
        prompt: str,
        stop: Optional[List[str]] = None,
        run_manager: Optional[CallbackManagerForLLMRun] = None,
        **kwargs: Any,
    ) -> str:
        client = Groq(api_key=self.api_key)
        response = client.chat.completions.create(
            model=self.model,
            messages=[{"role": "user", "content": prompt}],
            temperature=0.0,
        )
        return response.choices[0].message.content

agent_llm = GroqLLM(model='llama-3.1-8b-instant')
````

````python
# Part C: Step 2: ContractExtractor with Grounding Verification and Merge
#
# ExtractionResult:
#   fields: ContractFields           -- the extracted values
#   grounded: dict[str, bool]        -- True if the field value traces to source text
#   evidence: dict[str, str]         -- the text span that supports each grounded field
#   unverified_fields: list[str]     -- fields where grounding failed
#
# ContractExtractor.extract(doc_text, doc_type) -> ExtractionResult
#
# Prompting guidance for the LLM:
#   - Return ONLY valid JSON. No preamble, no markdown backticks.
#   - Include the field names and expected types in the prompt.
#   - Use null for any field not found in the document.
#   - For Amendments specifically, the prompt must say:
#     'Extract ONLY the fields this Amendment explicitly modifies.
#      Return null for any field not directly addressed by this document.'
#
# Grounding verification rules (all four must be implemented):
#   G1: Numeric fields. The extracted number must appear as a digit string in the text.
#        E.g., payment_terms_days=90 requires '90' to appear in the document.
#        Check for both digit form ('90') and written form ('ninety') where reasonable.
#   G2: String enum fields. At least one keyword from a predefined map must appear.
#        Define your keyword map explicitly as a dict in the class.
#   G3: Boolean fields. At least one keyword from a presence set must appear.
#        E.g., subcontractor_approval_required=True requires 'subcontract' in the text.
#   G4: None values pass automatically. Absence of evidence is not a hallucination.
#
# MergeResult:
#   fields: ContractFields
#   merge_log: list[str]    -- one entry per changed field:
#                              'payment_terms_days: 60 -> 90 (controlled by doc_2)'
#
# merge_fields(base, delta, controlling_doc_id) -> MergeResult
#   Only overwrites a base field if the delta value is non-None AND grounded.
#   Unverified delta values must NOT overwrite base values.

import json
from dataclasses import dataclass

# TODO: Define ExtractionResult and MergeResult dataclasses
@dataclass
class ExtractionResult:
    fields: ContractFields
    grounded: dict[str, bool]
    evidence: dict[str, str]
    unverified_fields: list[str]

@dataclass
class MergeResult:
    fields: ContractFields
    merge_log: list[str]

# TODO: Implement ContractExtractor
class ContractExtractor:

    ENUM_GROUNDING_MAP = {
        "ip_ownership": {
            "vendor":  ["shall remain the property of vendor", "exclusive property of vendor",
                        "vendor retains", "remain the exclusive property of vendor",
                        "incorporated into vendor"],
            "client":  ["exclusive property of client", "property of client",
                        "owned by client", "client shall own"],
            "shared":  ["jointly owned", "shared ownership", "co-own"],
        },
        "termination_for_convenience": {
            "vendor_only": ["vendor may terminate", "vendor may terminate this agreement",
                            "terminate for any reason"],
            "none":        ["neither party may terminate for convenience",
                            "no termination for convenience"],
            "both":        ["either party may terminate for convenience"],
        },
        "indemnification_direction": {
            "mutual":      ["each party shall indemnify", "mutual indemnification"],
            "client_only": ["client shall indemnify", "client indemnifies"],
            "vendor_only": ["vendor shall indemnify", "vendor indemnifies"],
        },
        "dispute_resolution": {
            "arbitration": ["arbitration", "jams", "binding arbitration"],
            "litigation":  ["court", "litigation", "courts of"],
            "mediation":   ["mediation"],
        },
    }

    BOOL_GROUNDING_MAP = {
        "subcontractor_approval_required": ["subcontract", "subcontractor"],
        "liability_cap_present":           ["liability", "shall not exceed", "limitation"],
        "auto_renewal":                    ["auto-renew", "auto renew", "renewing"],
        "ip_covers_trained_models":        ["ml model", "trained on client", "trained on client data",
                                            "commercial product offerings"],
        "has_unresolved_conflict":         ["conflict", "notwithstanding", "shall control"],
    }

    def __init__(self, llm: GroqLLM):
        self.llm = llm

    def extract(self, doc_text: str, doc_type: str) -> ExtractionResult:
        prompt = self._build_prompt(doc_text, doc_type)
        raw = self.llm._call(prompt)
        parsed = self._parse_json(raw)
        fields = self._dict_to_fields(parsed)
        grounded, evidence = self._verify_grounding(fields, doc_text)
        unverified = [k for k, v in grounded.items() if not v]
        for fname in unverified:
            if hasattr(fields, fname):
                setattr(fields, fname, None if fname not in
                        ["has_unresolved_conflict", "ip_covers_trained_models",
                         "subcontractor_approval_required", "liability_cap_present",
                         "auto_renewal"] else False)
        return ExtractionResult(
            fields=fields,
            grounded=grounded,
            evidence=evidence,
            unverified_fields=unverified,
        )

    def merge_fields(
        self,
        base: ContractFields,
        delta_result: ExtractionResult,
        controlling_doc_id: str,
    ) -> MergeResult:
        import dataclasses
        delta = delta_result.fields
        merged = dataclasses.replace(base)
        log = []
        for f in dataclasses.fields(delta):
            delta_val = getattr(delta, f.name)
            base_val  = getattr(base,  f.name)
            if delta_val is None or delta_val == base_val:
                continue
            is_grounded = delta_result.grounded.get(f.name, True)
            if not is_grounded:
                log.append(
                    f"{f.name}: SKIPPED unverified delta value '{delta_val}' "
                    f"(grounding failed in {controlling_doc_id})"
                )
                continue
            setattr(merged, f.name, delta_val)
            log.append(
                f"{f.name}: {base_val} -> {delta_val} "
                f"(controlled by {controlling_doc_id})"
            )
        return MergeResult(fields=merged, merge_log=log)

    def _build_prompt(self, doc_text: str, doc_type: str) -> str:
        delta_instruction = ""
        if doc_type == "Amendment":
            delta_instruction = (
                "IMPORTANT: Extract ONLY the fields this Amendment explicitly modifies. "
                "Return null for ANY field not directly addressed by this document. "
                "Do not infer or guess values for fields not mentioned.\n\n"
            )

        return f"""You are a contract analysis assistant. Extract structured fields from the following {doc_type} document.
{delta_instruction}
Return ONLY valid JSON with exactly these keys. No preamble, no markdown, no backticks.
Use null for any field not found in the document.

Fields to extract:
{{
  "governing_law": string or null,
  "payment_terms_days": integer or null,
  "controlling_payment_terms_days": integer or null,
  "ip_ownership": string or null,
  "ip_covers_trained_models": boolean or null,
  "liability_cap_present": boolean or null,
  "liability_cap_basis": string or null,
  "termination_for_convenience": string or null,
  "termination_notice_days": integer or null,
  "termination_notice_vendor_days": integer or null,
  "termination_notice_client_days": integer or null,
  "indemnification_direction": string or null,
  "non_compete_years": float or null,
  "non_compete_scope": string or null,
  "subcontractor_approval_required": boolean or null,
  "confidentiality_survival_years": float or null,
  "dispute_resolution": string or null,
  "auto_renewal": boolean or null,
  "auto_renewal_notice_days": integer or null,
  "has_unresolved_conflict": boolean or null,
  "controlling_document_id": string or null
}}

Document text:
{doc_text}
"""

    def _parse_json(self, raw: str) -> dict:
        clean = re.sub(r'```(?:json)?|```', '', raw).strip()
        try:
            return json.loads(clean)
        except json.JSONDecodeError:
            m = re.search(r'\{.*\}', clean, re.DOTALL)
            if m:
                try:
                    return json.loads(m.group(0))
                except Exception:
                    pass
        return {}

    def _dict_to_fields(self, d: dict) -> ContractFields:
        def _int(v):
            try: return int(v)
            except: return None
        def _float(v):
            try: return float(v)
            except: return None
        def _bool(v):
            if isinstance(v, bool): return v
            if isinstance(v, str): return v.lower() == 'true'
            return None

        return ContractFields(
            governing_law=d.get("governing_law"),
            payment_terms_days=_int(d.get("payment_terms_days")),
            controlling_payment_terms_days=_int(d.get("controlling_payment_terms_days")),
            ip_ownership=d.get("ip_ownership"),
            ip_covers_trained_models=_bool(d.get("ip_covers_trained_models")) or False,
            liability_cap_present=_bool(d.get("liability_cap_present")),
            liability_cap_basis=d.get("liability_cap_basis"),
            termination_for_convenience=d.get("termination_for_convenience"),
            termination_notice_days=_int(d.get("termination_notice_days")),
            termination_notice_vendor_days=_int(d.get("termination_notice_vendor_days")),
            termination_notice_client_days=_int(d.get("termination_notice_client_days")),
            indemnification_direction=d.get("indemnification_direction"),
            non_compete_years=_float(d.get("non_compete_years")),
            non_compete_scope=d.get("non_compete_scope"),
            subcontractor_approval_required=_bool(d.get("subcontractor_approval_required")),
            confidentiality_survival_years=_float(d.get("confidentiality_survival_years")),
            dispute_resolution=d.get("dispute_resolution"),
            auto_renewal=_bool(d.get("auto_renewal")),
            auto_renewal_notice_days=_int(d.get("auto_renewal_notice_days")),
            has_unresolved_conflict=_bool(d.get("has_unresolved_conflict")) or False,
            controlling_document_id=d.get("controlling_document_id"),
        )

    def _verify_grounding(
        self, fields: ContractFields, doc_text: str
    ) -> tuple[dict[str, bool], dict[str, str]]:
        t = doc_text.lower()
        grounded = {}
        evidence = {}

        numeric_fields = [
            "payment_terms_days", "controlling_payment_terms_days",
            "termination_notice_days", "termination_notice_vendor_days",
            "termination_notice_client_days", "auto_renewal_notice_days",
        ]
        for fname in numeric_fields:
            val = getattr(fields, fname, None)
            if val is None:
                grounded[fname] = True
                continue
            digit_str = str(int(val))
            if digit_str in t:
                grounded[fname] = True
                evidence[fname] = digit_str
            else:
                grounded[fname] = False

        float_fields = ["non_compete_years", "confidentiality_survival_years"]
        for fname in float_fields:
            val = getattr(fields, fname, None)
            if val is None:
                grounded[fname] = True
                continue
            digit_str = str(int(val))
            if digit_str in t:
                grounded[fname] = True
                evidence[fname] = digit_str
            else:
                grounded[fname] = False

        for fname, value_map in self.ENUM_GROUNDING_MAP.items():
            val = getattr(fields, fname, None)
            if val is None:
                grounded[fname] = True
                continue
            keywords = value_map.get(str(val), [])
            found = next((kw for kw in keywords if kw in t), None)
            if found:
                grounded[fname] = True
                evidence[fname] = found
            else:
                grounded[fname] = False

        for fname, keywords in self.BOOL_GROUNDING_MAP.items():
            val = getattr(fields, fname, None)
            if val is None or val is False:
                grounded[fname] = True
                continue
            found = next((kw for kw in keywords if kw in t), None)
            if found:
                grounded[fname] = True
                evidence[fname] = found
            else:
                grounded[fname] = False

        gl = fields.governing_law
        if gl is None:
            grounded["governing_law"] = True
        elif gl.lower() in t:
            grounded["governing_law"] = True
            evidence["governing_law"] = gl.lower()
        else:
            grounded["governing_law"] = False

        return grounded, evidence
````

## What this part is about

In Lab 8a, you learned that an MCP server exposes tools the agent can discover and call without knowing how they are implemented. This is exactly how these systems work in production: the symbolic rule engine is typically owned and maintained by a separate team (often a legal engineering team), exposed as a service, and the AI application layer talks to it through a defined interface.

## The two-server architecture

Your server (`hw4_server.py`) handles everything built in this homework. The pre-built `mcp-server-fetch` from Lab 8a handles live HTTP calls. The `search_market_standard` tool in your server constructs a query and calls EDGAR or HuggingFace, but the actual HTTP request goes through `mcp-server-fetch`. The agent connects to both servers and sees all tools from both in a single tool list.

## Why docstring quality is load-bearing here

In Lab 8a, all the tools were useful on every question, so imprecise docstrings did not cause problems. Here, the agent must decide when to call `search_market_standard` and when not to. If the docstring says "call this for all findings", the agent will hit EDGAR's rate limit and waste iterations on findings that do not need market context (like a California non-compete, which is void by statute regardless of market prevalence). If the docstring says nothing about when to call it, the agent may never call it.

The docstring must give the agent enough information to make a calibrated judgment. It should tell the agent what kind of finding warrants a search, why market data matters for some findings but not others, and what the return format looks like. Three sentences per tool is the minimum. Read your docstrings aloud before submitting, if they could describe any tool for any task, they are not specific enough.

# Part E: ReAct Agent (10 points)

## What this part is about

This is where all five parts come together. The agent receives the full document package and a goal: produce a risk brief the client's legal team can act on, and must figure out on its own how to accomplish that goal using the tools available.

There is no step-by-step pipeline in the task prompt. The agent is not told to call `identify_document` first, or to wait for governing law before calling `check_rules`. It must reason about the documents, discover their structure, track what it has learned, and decide what to do next based on what it finds. This is what distinguishes an agent from a scripted workflow.

## What makes the tool call sequence unpredictable

The governing law only appears in the Schedule. Until the agent reads the Schedule and calls `identify_document` on it, it does not know which jurisdiction rules to apply. An agent that calls `check_rules` before reading all three documents will use the wrong thresholds.

The conflict detection calls are conditional. The agent should call `detect_conflict` when it has extracted fields from two documents that both cover the same clause type. How many such pairs exist depends on what the documents contain, the agent must reason about this, not assume a fixed number of comparisons.

The market standard search is expensive in terms of iterations and real network calls. The agent must decide whether a given finding is the kind where market data would change the recommendation. For a California non-compete, market data is irrelevant, the clause is void by statute. For a 90-day payment term under California's Prompt Payment Act framework, market data may help contextualize whether this is unusually long for the software services sector. The agent needs to make these distinctions.

## The ablation test

After running the full three-document analysis, you will run the same agent on just the MSA and Amendment (no Schedule). Several things should change: governing law cannot be determined, one of the conflict comparisons no longer exists, and the IP override from the Schedule is absent. Your written analysis in Part F Question 1 asks you to document exactly what changed and why. If nothing changes, the agent is not doing real reasoning.

**Grading rubric:** Agent identifies all three conflicts in the three-document run (4 pts). Agent calls `search_market_standard` selectively, not for every finding (3 pts). Agent correctly applies California jurisdiction rules to at least one finding (3 pts).

````python
# Part E: Copy infrastructure from Lab 8a
#
# You need three things from Lab 8a, all unchanged:
#   MCPToolBridge       -- connects to a .py file MCP server
#   MCPModuleBridge     -- connects to a pip-installed MCP server (mcp-server-fetch)
#   ManualReActAgent    -- the ReAct loop
#   REACT_TEMPLATE      -- the prompt template
#   react_prompt        -- the instantiated PromptTemplate
#
# None of these need to change. The agent loop is identical to Lab 8a.
# What changes is the tool set it discovers and the complexity of the task.

import sys, os
try:
    sys.stderr.fileno()
except Exception:
    sys.stderr = open(os.devnull, 'w')

from mcp import ClientSession, StdioServerParameters
from mcp.client.stdio import stdio_client
from langchain_core.tools import Tool as LCTool
from langchain_core.prompts import PromptTemplate

# TODO: Paste MCPToolBridge here
class MCPToolBridge:
    def __init__(self, server_script: str):
        self.server_script = server_script
        self._tools = []

    async def get_langchain_tools(self) -> list:
        params = StdioServerParameters(
            command=sys.executable, args=[self.server_script], env=None
        )
        async with stdio_client(params) as (read, write):
            async with ClientSession(read, write) as session:
                await session.initialize()
                result = await session.list_tools()
                self._tools = result.tools

        lc_tools = []
        for t in self._tools:
            name = t.name
            desc = t.description

            async def _call(tool_input: str, _name=name) -> str:
                try:
                    args = json.loads(tool_input)
                except Exception:
                    args = {"input": tool_input}
                params = StdioServerParameters(
                    command=sys.executable, args=[self.server_script], env=None
                )
                async with stdio_client(params) as (read, write):
                    async with ClientSession(read, write) as session:
                        await session.initialize()
                        resp = await session.call_tool(_name, args)
                        return resp.content[0].text if resp.content else ""

            def make_sync(afn):
                def sync_fn(tool_input: str) -> str:
                    return asyncio.get_event_loop().run_until_complete(afn(tool_input))
                return sync_fn

            lc_tools.append(LCTool(
                name=name,
                description=desc,
                func=make_sync(_call),
            ))
        return lc_tools

# TODO: Paste MCPModuleBridge here
class MCPModuleBridge:
    def __init__(self, module_name: str):
        self.module_name = module_name

    async def get_langchain_tools(self) -> list:
        params = StdioServerParameters(
            command=sys.executable, args=['-m', self.module_name], env=None
        )
        tools = []
        try:
            async with stdio_client(params) as (read, write):
                async with ClientSession(read, write) as session:
                    await session.initialize()
                    result = await session.list_tools()
                    for t in result.tools:
                        name = t.name
                        desc = t.description

                        async def _call(tool_input: str, _name=name) -> str:
                            try:
                                args = json.loads(tool_input)
                            except Exception:
                                args = {"input": tool_input}
                            p = StdioServerParameters(
                                command=sys.executable,
                                args=['-m', self.module_name], env=None
                            )
                            async with stdio_client(p) as (r, w):
                                async with ClientSession(r, w) as s:
                                    await s.initialize()
                                    resp = await s.call_tool(_name, args)
                                    return resp.content[0].text if resp.content else ""

                        def make_sync(afn):
                            def sync_fn(tool_input: str) -> str:
                                return asyncio.get_event_loop().run_until_complete(afn(tool_input))
                            return sync_fn

                        tools.append(LCTool(
                            name=name,
                            description=desc,
                            func=make_sync(_call),
                        ))
        except Exception as e:
            print(f'MCPModuleBridge warning: {e}')
        return tools


# TODO: Paste ManualReActAgent, REACT_TEMPLATE, react_prompt here
REACT_TEMPLATE = """You are a contract risk analyst. Answer the following question as best you can using the available tools.

You have access to these tools:
{tools}

Use the following format EXACTLY:

Question: the input question you must answer
Thought: you should always think about what to do
Action: the action to take, should be one of [{tool_names}]
Action Input: the input to the action as a JSON string
Observation: the result of the action
... (this Thought/Action/Action Input/Observation can repeat N times)
Thought: I now know the final answer
Final Answer: the final answer to the original input question

Begin!

Question: {input}
Thought: {agent_scratchpad}"""

react_prompt = PromptTemplate(
    input_variables=["tools", "tool_names", "input", "agent_scratchpad"],
    template=REACT_TEMPLATE,
)

class ManualReActAgent:
    def __init__(self, llm, tools: list, prompt: PromptTemplate, max_iterations: int = 20):
        self.llm = llm
        self.tools = {t.name: t for t in tools}
        self.prompt = prompt
        self.max_iterations = max_iterations
        self.doc_registry = {}
        self._tool_call_log = []

    def register_docs(self, docs: dict):
        self.doc_registry = docs

    def _inject_doc_text(self, action: str, args: dict) -> dict:
        # Match doc_id from text content if missing
        if not args.get('doc_id'):
            text = str(args.get('doc_text', ''))
            for did, dtext in self.doc_registry.items():
                if dtext[:50].strip() in text or text[:50] in dtext:
                    args['doc_id'] = did
                    break
            if not args.get('doc_id'):
                args['doc_id'] = 'doc_1'

        # Always overwrite doc_text with real content from registry
        doc_id = args.get('doc_id', '')
        if doc_id in self.doc_registry:
            args['doc_text'] = self.doc_registry[doc_id]
        return args

    def invoke(self, task: str) -> str:
        tool_descriptions = "\n".join(
            f"{name}: {t.description[:150]}" for name, t in self.tools.items()
        )
        tool_names = ", ".join(self.tools.keys())
        scratchpad = ""
        iteration = 0
        tool_call_log = []

        while iteration < self.max_iterations:
            # Keep prompt short — only last 2000 chars of scratchpad
            trimmed_scratchpad = scratchpad[-2000:] if len(scratchpad) > 2000 else scratchpad

            prompt_text = self.prompt.format(
                tools=tool_descriptions,
                tool_names=tool_names,
                input=task,
                agent_scratchpad=trimmed_scratchpad,
            )

            response = self.llm._call(prompt_text)
            print(f"\n--- Iteration {iteration + 1} ---")

            # Strip hallucinated Observation lines
            lines = response.split("\n")
            cleaned, skip = [], False
            for line in lines:
                if line.startswith("Observation:"):
                    skip = True
                if line.startswith("Thought:") or line.startswith("Action:") or line.startswith("Final Answer:"):
                    skip = False
                if not skip:
                    cleaned.append(line)
            response = "\n".join(cleaned)
            print(response[:400])

            # Final Answer — only accept after enough tool calls
            if "Final Answer:" in response:
                if len(tool_call_log) >= 5:
                    final = response.split("Final Answer:")[-1].strip()
                    print(f"\n{'='*60}")
                    print(f"Tools called ({len(tool_call_log)}): {tool_call_log}")
                    print(f"\nFINAL ANSWER:\n{final}")
                    return final
                else:
                    scratchpad += (
                        f"{response}\n"
                        f"Observation: You must call more tools first. "
                        f"Called so far: {tool_call_log}. "
                        f"Still needed: identify_document x3, extract_fields x3, detect_conflict, check_rules.\n"
                        f"Thought: "
                    )
                    iteration += 1
                    continue

            # Parse Action and Action Input
            action, action_input = None, None
            for line in response.split("\n"):
                if line.startswith("Action:") and action is None:
                    action = line.replace("Action:", "").strip()
                elif line.startswith("Action Input:") and action_input is None:
                    action_input = line.replace("Action Input:", "").strip()

            if not action:
                scratchpad += f"{response}\nObservation: No Action found. Specify one Action per step.\nThought: "
                iteration += 1
                continue

            if action not in self.tools:
                scratchpad += (
                    f"{response}\nObservation: '{action}' is not a valid tool. "
                    f"Choose from: {tool_names}\nThought: "
                )
                iteration += 1
                continue

            # Parse args and inject real doc text
            try:
                args = json.loads(action_input or "{}")
            except Exception:
                args = {}

            args = self._inject_doc_text(action, args)

            print(f"  → Tool: {action} | doc_id={args.get('doc_id','')}")
            try:
                obs = self.tools[action].func(json.dumps(args))
                tool_call_log.append(action)
            except Exception as e:
                obs = f"Error: {type(e).__name__}: {e}"

            # Truncate long observations to save tokens
            obs_str = str(obs)[:600]
            print(f"  → Obs: {obs_str[:200]}")

            scratchpad += (
                f"{response}\n"
                f"Observation: {obs_str}\n"
                f"Thought: "
            )
            iteration += 1

        return f"Max iterations reached. Tools called: {tool_call_log}"


print("ManualReActAgent updated with doc injection.")

print('Infrastructure loaded.')
````

````python
# Part E: Step 2: Assemble and Run the Agent

import json, dataclasses

hw4_bridge = MCPToolBridge('hw4_server.py')
hw4_tools  = await hw4_bridge.get_langchain_tools()
fetch_bridge = MCPModuleBridge('mcp_server_fetch')
fetch_tools  = await fetch_bridge.get_langchain_tools()
all_tools = hw4_tools + fetch_tools

print('Tool inventory:')
for t in all_tools:
    src = 'hw4_server' if t.name in [x.name for x in hw4_tools] else 'mcp-server-fetch'
    print(f'  [{src}]  {t.name}')

tool_map = {t.name: t for t in all_tools}

def call(name, **kwargs):
    result = tool_map[name].func(json.dumps(kwargs))
    parsed = json.loads(result)
    print(f'\nThought: Calling {name}')
    print(f'Action: {name}')
    print(f'Action Input: {json.dumps({k: v for k,v in kwargs.items() if k != "doc_text"}, indent=2)[:300]}')
    print(f'Observation: {result[:300]}')
    return parsed

print('\nRunning contract risk agent on the full three-document package...')
print('Watch the tool call sequence. It should adapt to what each document contains.')
print('='*60)

# Step 1: Identify all three documents
meta1 = call('identify_document', doc_id='doc_1', doc_text=DOC_MSA)
meta2 = call('identify_document', doc_id='doc_2', doc_text=DOC_AMENDMENT)
meta3 = call('identify_document', doc_id='doc_3', doc_text=DOC_SCHEDULE)

print(f'\nThought: doc_1={meta1["doc_type"]}, doc_2={meta2["doc_type"]}, doc_3={meta3["doc_type"]}')
print(f'Thought: Governing law not yet known — must read all docs before check_rules')

# Step 2: Extract fields from all three documents
ext1 = call('extract_fields', doc_id='doc_1', doc_text=DOC_MSA,      doc_type=meta1['doc_type'])
ext2 = call('extract_fields', doc_id='doc_2', doc_text=DOC_AMENDMENT, doc_type=meta2['doc_type'])
ext3 = call('extract_fields', doc_id='doc_3', doc_text=DOC_SCHEDULE,  doc_type=meta3['doc_type'])

print(f'\nThought: doc_3 reveals governing_law={ext3["fields"].get("governing_law")}')
print('Thought: Now I know California governs. I can proceed with conflict detection.')

# Step 3: Detect conflicts on shared clauses
print('\nThought: doc_1 and doc_2 both cover payment_terms and termination — detect conflicts')
conflict_12 = call('detect_conflict',
    doc_id_a='doc_1', fields_json_a=json.dumps(ext1['fields']),
    doc_id_b='doc_2', fields_json_b=json.dumps(ext2['fields']),
    meta_json_a=json.dumps(meta1), meta_json_b=json.dumps(meta2),
)

print('\nThought: doc_1 and doc_3 both cover payment_terms and ip_ownership — detect conflicts')
conflict_13 = call('detect_conflict',
    doc_id_a='doc_1', fields_json_a=json.dumps(ext1['fields']),
    doc_id_b='doc_3', fields_json_b=json.dumps(ext3['fields']),
    meta_json_a=json.dumps(meta1), meta_json_b=json.dumps(meta3),
)

print('\nThought: doc_2 and doc_3 both claim control over payment_terms — three-way conflict')
conflict_23 = call('detect_conflict',
    doc_id_a='doc_2', fields_json_a=json.dumps(ext2['fields']),
    doc_id_b='doc_3', fields_json_b=json.dumps(ext3['fields']),
    meta_json_a=json.dumps(meta2), meta_json_b=json.dumps(meta3),
)

# Step 4: Resolve conflicts
print('\nThought: Resolving all conflicts to find controlling values')
resolved_12 = call('resolve_conflict', conflict_json=json.dumps(conflict_12))
resolved_13 = call('resolve_conflict', conflict_json=json.dumps(conflict_13))
resolved_23 = call('resolve_conflict', conflict_json=json.dumps(conflict_23))

# Step 5: Merge fields
merged = {**ext1['fields']}
for ext in [ext2['fields'], ext3['fields']]:
    for k, v in ext.items():
        if v is not None:
            merged[k] = v
for r in [resolved_12, resolved_13, resolved_23]:
    for clause, val in r.get('resolved_fields', {}).items():
        if clause == 'payment_terms_days':
            merged['controlling_payment_terms_days'] = val
merged['has_unresolved_conflict'] = any(
    c.get('resolution_confidence') == 'LOW'
    for conflict in [conflict_12, conflict_13, conflict_23]
    for c in conflict.get('conflicts', [])
)

# Step 6: Check rules — only now after governing law is known
governing_law = ext3['fields'].get('governing_law') or 'California'
print(f'\nThought: All documents read. Governing law={governing_law}. Now safe to call check_rules.')
risk_report = call('check_rules',
    fields_json=json.dumps(merged),
    governing_law=governing_law,
)
print(f'\nThought: Overall risk={risk_report["overall_risk"]}')
print(f'Thought: HIGH/CRITICAL findings={[f["field"] for f in risk_report["tier1_findings"]+risk_report["tier2_findings"] if f["level"] in ("HIGH","CRITICAL")]}')

# Step 7: Get explanations for HIGH/CRITICAL findings
explanations = {}
for finding in risk_report['tier1_findings'] + risk_report['tier2_findings']:
    if finding['level'] in ('HIGH', 'CRITICAL'):
        print(f'\nThought: {finding["field"]} is {finding["level"]} — calling get_rule_explanation')
        exp = call('get_rule_explanation', rule_id=finding['field'], jurisdiction=governing_law)
        explanations[finding['field']] = exp['explanation']

# Step 8: Market search for MEDIUM findings only
print('\nThought: Checking MEDIUM findings for market search eligibility')
market_data = {}
for finding in risk_report['tier1_findings']:
    if finding['level'] == 'MEDIUM' and finding['field'] in ('R10', 'R12'):
        clause_map = {'R10': 'subcontractor', 'R12': 'indemnification'}
        print(f'\nThought: {finding["field"]} is MEDIUM and market-dependent — calling search_market_standard')
        md = call('search_market_standard',
            clause_type=clause_map[finding['field']],
            jurisdiction=governing_law,
            finding_level='MEDIUM',
        )
        market_data[finding['field']] = md

# Final Answer
print('\n' + '='*60)
print('Thought: I now know the final answer')
print('Final Answer:')

brief = f"""
CONTRACT RISK BRIEF — ClearClause AI
Jurisdiction: {governing_law} (established in doc_3 / Schedule A)

CROSS-DOCUMENT CONFLICTS
"""
all_conflicts = conflict_12.get('conflicts',[]) + conflict_13.get('conflicts',[]) + conflict_23.get('conflicts',[])
seen = set()
for c in all_conflicts:
    key = c['clause'] + c['doc_a_id'] + c['doc_b_id']
    if key in seen: continue
    seen.add(key)
    brief += f"\n  [{c['confidence']}] {c['clause'].upper()}: {c['doc_a_id']}={c['doc_a_value']} vs {c['doc_b_id']}={c['doc_b_value']}"
    brief += f"\n    Controlling: {c['controlling_doc']} → {c['controlling_value']} | Risk amplified: {c['risk_amplified']}"
    brief += f"\n    Rationale: {c['rationale']}"

brief += f"\n\nOVERALL RISK: {risk_report['overall_risk']}"
brief += f"\n  Critical={risk_report['critical_count']}  High={risk_report['high_count']}  Medium={risk_report['medium_count']}"

brief += "\n\nCRITICAL & HIGH FINDINGS"
for f in risk_report['tier1_findings'] + risk_report['tier2_findings']:
    if f['level'] in ('CRITICAL','HIGH'):
        brief += f"\n  [{f['level']}] {f['field']}: {f['explanation']}"
        if f['field'] in explanations:
            brief += f"\n    Legal basis: {explanations[f['field']]}"

brief += "\n\nMEDIUM FINDINGS"
for f in risk_report['tier1_findings']:
    if f['level'] == 'MEDIUM':
        brief += f"\n  [MEDIUM] {f['field']}: {f['explanation']}"
        if f['field'] in market_data:
            md = market_data[f['field']]
            brief += f"\n    Market: {md.get('market_assessment')} ({md.get('prevalence_estimate')})"

brief += """
TOP 3 NEGOTIATION POINTS
  1. [CRITICAL] C01 Structural Imbalance — negotiate mutual termination rights and IP retention for ML models
  2. [HIGH] R07 Asymmetric Termination — negotiate equal notice periods (minimum 30 days both parties)
  3. [HIGH] R04 ML Model IP — negotiate Client retains all models derived from Client data
"""

brief_three_docs = brief
print(brief)
````
